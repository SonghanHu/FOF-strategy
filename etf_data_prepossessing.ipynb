```python
# Convert index to datetime if it's not already
df_all.index = pd.to_datetime(df_all.index)

# Calculate average volumes and sort them
average_volumes = df_all.groupby('code')['volume'].mean()
average_volumes = average_volumes.sort_values()

# Exclude ETFs with the 100 lowest average volumes
df = df_all[~df_all['code'].isin(average_volumes[:150].index)]
print(len(df['code'].unique()))
# Exclude ETFs with less than one year of data
#years_for_each_etf = df.groupby('code').apply(lambda x: x.index.year.unique())
#etfs_less_than_one_year = years_for_each_etf[years_for_each_etf.apply(len) < 1].index
#df = df[~df['code'].isin(etfs_less_than_one_year)]
#print(len(df['code'].unique()))
# Calculate daily returns
df['return'] = df.groupby('code')['close'].pct_change()

# Create a pivot table and calculate the correlation matrix
df_pivot = df.pivot(columns='code', values='return')
correlation_matrix = df_pivot.corr()

# Exclude ETFs with high correlations
etfs_to_drop = []

for i in range(correlation_matrix.shape[0]):
    for j in range(i+1, correlation_matrix.shape[1]):
        if correlation_matrix.iloc[i, j] > 0.93:
            # change the correlation threshold here
            if average_volumes[correlation_matrix.columns[i]] > average_volumes[correlation_matrix.columns[j]]:
                etfs_to_drop.append(correlation_matrix.columns[j])
            else:
                etfs_to_drop.append(correlation_matrix.columns[i])

df = df[~df['code'].isin(etfs_to_drop)]

print(len(df['code'].unique()))

# Calculate annual returns for each ETF
df['year'] = df.index.year
annual_returns = df.groupby(['code', 'year'])['return'].apply(lambda x: (1 + x).prod() - 1)
annual_returns = annual_returns.reset_index()

# Calculate annual variance for each ETF
annual_variances = annual_returns.groupby('code')['return'].var()

# Exclude the ETFs with the top 10 highest variances
sorted_variances = annual_variances.sort_values(ascending=False)
etfs_to_drop = sorted_variances.head(10).index
df = df[~df['code'].isin(etfs_to_drop)]
code = df.code.unique()
print(len(df['code'].unique()))
```
